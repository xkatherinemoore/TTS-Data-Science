```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

company_sales = pd.read_csv("data/company_sales_data.csv")
company_sales.head()
```

# Exercise 1
Read Total profit of all months and show it using a line plot

```python
plt.plot(company_sales['month_number'], company_sales['total_profit'])
plt.xlabel('Month')
plt.ylabel('Total Profit ($)')
plt.title('Total Profit by Month')
plt.axis((1, 12, 150000, 450000))
plt.grid()
plt.show()
```

# Exercise 2
Plot Total Profit of each month with a line plot; use the given styling parameters

```python
plt.title('Total Profit by Month')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')

plt.plot(company_sales['month_number'], company_sales['total_profit'], label='Total Profit Over Time', linestyle='dotted', color='red', marker='o', markerfacecolor='red', linewidth = 3)

plt.legend(loc='lower right')
plt.axis((1, 12, 150000, 450000))
plt.grid()

plt.show()
```

# Exercise 6
Create a script that creates either a vector or raster file format, and display the visual within the notebook. Discuss why you prefer either vector or raster image files.

```python
def save_file(file_name):
    file = file_name + '.pdf'
    plt.savefig(file)

save_file('face_products')

# Requires PyPDF2 to read pdf in python
import PyPDF2

reader = PyPDF2.PdfReader('face_products.pdf')
page = reader.pages[0]
count = 0

# Extracts images from pdfs
for image_file_obj in page.images:
    with open(str(count) +image_file_obj.name, "wb") as fp:
        fp.write(image_file_obj.data)
        count +=1
```

In most instances, I prefer vector image files over raster files because they are not pixelated. For large amounts of data, pixelation can make the visual difficult to read and even potentially misleading. The other benefit of vector images is that they are scalable. Because they are not made up of pixels, they can be zoomed in or out without any distortion to the image.
