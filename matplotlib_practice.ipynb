# Matplotlib Basics

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 1. Draw a line
x = range(0, 100)
y = [value for value in x]
plt.plot(x, y)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Draw a line')

plt.show()
```

```python
# 2. Plot specific points
x = [1.0, 2.0, 3.0]
y = [2.0, 4.0, 1.0]
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axis((1.0, 3.0, 1.0, 4.0))
plt.title('Sample Graph')
```

```python
# 3. Import a test file and graph
data = pd.read_csv("./data/test.txt", delimiter=" ", header=None)
plt.plot(data[0], data[1])

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Test Data')
plt.axis((1, 3, 1, 4))
```

```python
# 4. Draw line charts of multiple data streams
google_fdata = pd.read_csv("./data/fdata.csv", delimiter=",", index_col="Date")
google_fdata.plot(title="Alphabet Financial Data")

```

```python
# 5. Plot two or more lines with legend
x = [10, 20, 30]
line1 = [20, 40, 10]
line2 = [40, 10, 30]
plt.plot(x, line1, label = "line 1")
plt.plot(x, line2, label="line 2")

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two Lines')
plt.legend()
```

```python
# 6. Format graphs
plt.plot(x, line1, label="line1-width-3", color="blue", linewidth = 3)
plt.plot(x, line2, label="line2-width-5", color="red", linewidth = 5)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two Lines, Different Widths')
plt.legend()
```

```python
# 7. Plot two different styles
plt.plot(x, line1, label="line1-dotted", color="blue", linestyle="dotted")
plt.plot(x, line2, label="line2-dashed", color="red", linestyle="dashed")

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two Lines, Different Styles')
plt.legend()
```

```python
# 8. Plot different line markers
x = [1, 4, 5, 6, 7]
y = [2, 6, 3, 6, 3]
plt.plot(x, y, color="red", linestyle="dashdot", linewidth = 3, marker='o', markerfacecolor='blue', markersize = 12)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two Lines, Different Widths')
plt.axis((1, 8, 1, 8))
```
